The dataset contains cases from a study that was conducted between 1958 and 1970 at the University of Chicago's Billings Hospital on the survival of patients who had undergone surgery for breast cancer.

*Finding out, how many patients survived the cancer after the year of surgery
let us define and understand the variables first.

1. Age: Age of the patient at the time of study
2. Year: Year of operation of the patient.
3. Nodes: The lymph nodes are small bean like organs found in the human body. Preserence of cancerous cells in the lymph nodes under the arm may suggest that there might be possibility of breast cancer in the patient.
4. Status: The status indicates survival rate of the patient who has been through surgery and has surviced more than 5 years. In the data set, the number '1' indicates that the patient survived the surgery and '2' indicates that the patient died within 5 years of surgery.

# Univariate Analysis

```python
haberman_alive = haberman.loc[haberman["status"] == 1]
haberman_alive
#creating a seperate dataset for patients who have survived more than 5 years
```

```python
haberman_dead = haberman.loc[haberman["status"] == 2]
haberman_dead
#dataset containing patients who died within 5 years or undergoing operation
```

```python
plt.plot(haberman_alive["nodes"], np.zeros_like(haberman_alive["nodes"]))
plt.show()
```

```python
print("Univariate Analysis using histogram with Lymph node Data")
sns.FacetGrid(haberman, hue="status", height=5) \
   .map(sns.distplot, "nodes") \
   .add_legend();
plt.show();
```

From the above graph, which was plotted to see the difference and to understand that the numner of nodes present in the patients. However, by plotting this graph, precisely, a histogram, we can conclude that most cancer patients who were selected for study, had nodes close to 1. It is also to be noted that the data label indicates 1, meaning the patient lived and 2 indicating the patient died within 5 years of surgery. It is also to be found that the death rate is significantly high when the nodes are above 5, as we can see the PDF crossover at 5. Meaning, the patients who had more than 5 nodes, although lived, but the patients who passed away are more.

```python
print("Univariate Analysis using histogram with respect to age")
sns.FacetGrid(haberman, hue="status", height=5) \
   .map(sns.distplot, "age") \
   .add_legend();
plt.show();
```

# Analysis of the above plot
The above histogram plot is indicating the distribution of patient's ages included in the study. Please note that the maximum woman who are a victim of breast cancer, vary inbetween the age of 40 to 45, who also are vulnerable and might experience death.

The plot also shows that the survival rate of patients between the age 50 and 60 show more survival rate than the death, however, it is inconclusive.

```python
print("Box Plot wrt status of patient vs age")
sns.boxplot(x='status',y='age', data=haberman)
plt.show()
```

```python
print("Box Plot wrt status of patient vs Year of operation")
sns.boxplot(x='status',y='year', data=haberman)
plt.show()
```

```python
print("Violin Plot wrt status of patient vs age")
sns.violinplot(x='status',y='age', data=haberman, size = 5)
plt.show()
```

1. The Survival rate is maximum at the age of approximately 53
2. The death rate is maximum at the age of 50

```python
print("Violin Plot wrt status of patient vs Lymph nodes")
sns.violinplot(x='status',y='nodes', data=haberman, size = 5)
plt.show()
```

1. The survival rate of pateints greatly depends on the number of Lymph nodes present. We can confirm using this violin plot.
2. 

```python
counts, bin_edges = np.histogram(haberman['age'], bins=10, density = True)

pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label = "PDF age")
plt.plot(bin_edges[1:], cdf, label = "CDF age")


counts, bin_edges = np.histogram(haberman['nodes'], bins=20, 
                                 density = True)
pdf = counts/(sum(counts))
plt.plot(bin_edges[1:],pdf, label = "PDF nodes");
plt.title("PDF and CDF based on Age of Patient")
plt.xlabel("Bin edges for nodes")
plt.ylabel("Density/Percentage")
plt.legend(loc = "upper left")

plt.show();
```

```python
counts, bin_edges = np.histogram(haberman_dead['nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print("PDF Haberman_dead: ",pdf)
print("Bin Edges: ", bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf, label = "PDF")
plt.plot(bin_edges[1:], cdf, label = "CDF")
plt.title("PDF and CDF based on Lymph node data")
plt.xlabel("Bin edges")
plt.ylabel("Auxillary Lymph Nodes Density")
plt.legend(loc = "upper left")
plt.show();
```

# Summary of Univariate Analysis
1. We can see that the death of patients is affected mainly by preserence of more number of Lymph nodes.
2. We can also see some relation with the age being a factor which will may affect the survival of a patient.

However, since there are no conclusive evidence of exact reason for death, we have to move forward with the next methods which may be able to identify what exactly causes the deaths.

# Bivariate Analysis

Bivariate analysis consists of analysis of data using two features and comparing how the features affect the survival rate of patients.

The following are the plots to be used to analyse the data:
1. scatter plots
2. pair-plots

```python
#2-D scatter plot:
haberman.plot(kind='scatter', x='age', y='nodes') ;
plt.show()
```

1. Cannot differentiate between the scatter points as to which patients are alive and which are not. 
2. We have to color the points by thier class-label/status-type.

# 2-D Scatter plot with color-coding for each status type/class.

```python
sns.set_style("whitegrid");
sns.FacetGrid(haberman, hue="status", height=4) \
   .map(plt.scatter, "age", "nodes") \
   .add_legend();
plt.show();
```

1. Notice that the blue points cannot be easily seperated from orange points by drawing a line.
2. However, we can see that the increase in nodes can significantly increase the rate of death, there are significantly more yellow points above the line where the nodes are more than 10, although there are points which are overlapping.

# Pair plots

```python
sns.set_style("whitegrid");
sns.pairplot(haberman, hue="status", height=5);
plt.show()
```

From the above Pair plots, we have obtained multiple graphs.
* The first plot to look at would be year vs age, which does not give any clear conclusions as the year of operations overlap due to the nature of data collected, hence we can move on to next pair plot.
* The second pair plot, which is plotted with nodes as x axis and age as y axis, gives a good spread of data. It indicates that the there are more chances of death of patient if the nodes are more than 5 and the age of the patient crosses the age of 45. This is the best plot which helps analyse the data. Since the nature of data depends greately on various factors, this is the best of analysis.

```python
#Mean, Variance, Std-deviation,  
print("Means:")
print(np.mean(haberman_alive["nodes"]))
print(np.mean(haberman_dead["nodes"]))

print("\nStd-dev:");
print(np.std(haberman_alive["nodes"]))
print(np.std(haberman_dead["nodes"]))
```
